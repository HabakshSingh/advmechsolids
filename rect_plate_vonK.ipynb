# Buckling of rectangular plate

Here, we consider the example problem given in Section 9.13 page 535 in _Solid Mechanics_ by Dym and Shames. A simply supported rectangular plate of length $a$, breadth $b$, and thickness $h$ is loaded compressively in its own plane; the compressive load is parallel to the side of length $a$. 

The following code generates Fig. 9.18. 

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 18,'text.usetex':True})

inc = 0.1
a_by_b = np.arange(inc,10+inc,inc)

m = 1
k_1 = (m/a_by_b + a_by_b/m)**2

fig, ax = plt.subplots(figsize=(6,6/1.62),nrows=1, ncols=1)    


ax.plot(a_by_b,k_1,'b--')
ax.set_ylim([2, 8])

m = 2
k_2 = (m/a_by_b + a_by_b/m)**2
ax.plot(a_by_b,k_2,'r--')

m = 3
k_3 = (m/a_by_b + a_by_b/m)**2
ax.plot(a_by_b,k_3,'g--')

m = 4
k_4 = (m/a_by_b + a_by_b/m)**2
ax.plot(a_by_b,k_4,'c--')

m = 5
k_5 = (m/a_by_b + a_by_b/m)**2
ax.plot(a_by_b,k_5,'m--')

m = 6
k_6 = (m/a_by_b + a_by_b/m)**2
ax.plot(a_by_b,k_6,'y--')


K = np.matrix([k_1,k_2,k_3,k_4,k_5,k_6])

kmin = np.zeros(len(a_by_b))
for j in range(0,len(a_by_b)):
    kmin[j] = np.min(K[:,j])
    
ax.plot(a_by_b,kmin,'k-')

ax.set_xlabel(r'$a/b$')
ax.set_ylabel(r'$\displaystyle \frac{1}{m}\frac{a}{b} + m\frac{b}{a}$')
```
